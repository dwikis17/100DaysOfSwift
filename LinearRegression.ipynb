```python
#import and initialize datasets
import pandas as pd
import numpy as np


data = pd.read_csv('Salary_Data.csv')
x_train = data[['YearsExperience']]
y_train = data[['Salary']]

x_train = np.array(x_train)
y_train = np.array(y_train)

w = 0
b = 0
```

```python
def compute_cost(x,y,w,b):
  m = x_train.shape[0]

  cost = 0
  total_cost = 0

  for i in range(m):
    f_wb = w * x[i] + b
    cost += f_wb - y[i]

  total_cost = cost ** 2

  return total_cost
```

```python
def gradient_function(x,y,w,b):
  m = x_train.shape[0]
  dj_dw = 0
  dj_db = 0


  for i in range(m):
    f_wb = w * x[i] + b
    dj_dw += (f_wb - y[i]) * x[i]
    dj_db += f_wb - y[i]

  return dj_dw/m, dj_db/m
```

```python
def gradient_descent(x,y,w,b):
  print(w,b)
  m = x_train.shape[0]
  alpha = 0.001
  iterations = 1000
  J_history = []

  for i in range(iterations):
    dj_dw, dj_db = gradient_function(x,y,w,b)
    w = w - alpha * dj_dw
    b = w - alpha * dj_db

  cost = compute_cost(x,y,w,b)
  J_history.append(cost)
  return w,b,J_history
```

```python
import matplotlib.pyplot as plt

w,b,J_history = gradient_descent(x_train, y_train, w, b)

m = x_train.shape[0]
prediction = np.zeros(m)

for i in range(m):
  prediction[i] = w * x_train[i] + b


plt.plot(x_train, prediction, c='b')
plt.title('Salary vs Years Experience')
plt.scatter(x_train, y_train, marker='x', c='r')
plt.xlabel("Work Experience in Year")
plt.ylabel('Salary')
plt.show()
predict1 = 0 * w + b
print("for 0 year of experience, this model will predict that the subject will get $%.2f" % (predict1))
```
